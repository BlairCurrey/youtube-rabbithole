```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
```

### Load and transform

```python
df = pd.read_json('results\\gold_recovery-500.json')
titles = df["title"].tolist()
```

### Build Similarity Matrix
adapted from here:
https://stackoverflow.com/questions/8897593/how-to-compute-the-similarity-between-two-text-documents

```python
vect = TfidfVectorizer(min_df=1, stop_words="english")
tfidf = vect.fit_transform(titles)
pairwise_similarity = tfidf * tfidf.T
```

```python
similarity = pairwise_similarity.toarray()
similarity[5][6]
```

### Show

0 is dissimilar, 1 is similar.

Index 0,0 shows how index 0 relates to index 0.
Index 0,2 show how the first title relates to the 3rd title.
Index 190,450 shows how the 191st title compares to the 451st title, etc.

```python
plt.rcParams['figure.figsize'] = [10, 10]
plt.imshow(similarity, cmap="copper", interpolation="nearest")
```
